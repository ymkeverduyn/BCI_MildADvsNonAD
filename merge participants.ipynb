## Merge Participants 

We have one csv file for each participant's resting state data 
We split the data into 18 epochs of 3 seconds each to match the ERP data 
For each epoch and each participant we have 230 features ectracted from the resting state: 
- for each of the 19 channels; for each of the 5 subbands; AmplitudeSummation and Variance 

- for each of the 8 channel pairs; for each of the 5 subbands; Coherence 

19x5x2 + 8x5x1 = 230 


```python
import pandas as pd 
import numpy as np 
```

### Do the following for each participant - chnage the paths according to the participant number 

```python
participant = pd.read_csv("New_EEG_Results_sub012\EEG_Features_Windows_sub012.csv")
participant.head()
```

```python
window_1 = participant.iloc[:, 0:230]
window_2 = participant.iloc[:, 230:460]
window_3 = participant.iloc[:, 460:690]
window_4 = participant.iloc[:, 690:920]
window_5 = participant.iloc[:, 920:1150]
window_6 = participant.iloc[:, 1150:1380]
window_7 = participant.iloc[:, 1380:1610]
window_8 = participant.iloc[:, 1610:1840]
window_9 = participant.iloc[:, 1840:2070]
window_10 = participant.iloc[:, 2070:2300]
window_11 = participant.iloc[:, 2300:2530]
window_12 = participant.iloc[:, 2530:2760]
window_13 = participant.iloc[:, 2760:2990]
window_14 = participant.iloc[:, 2990:3220]
window_15 = participant.iloc[:, 3220:3450]
window_16 = participant.iloc[:, 3450:3680]
window_17 = participant.iloc[:, 3680:3910]
window_18 = participant.iloc[:, 3910:4240]

display(window_1)
display(window_10)
```

```python
windows = [window_1, window_2, window_3, window_4, window_5, window_6, window_7, window_8, window_9, window_10, window_11,
           window_12, window_13, window_14, window_15, window_16, window_17, window_18]

all_epochs = pd.DataFrame()

for window in windows: 
    window = window.rename(lambda x: '_'.join(x.split('_')[1:]), axis=1)

    #display(window)

    all_epochs = pd.concat([all_epochs, window], axis=0)
    #all_epochs = pd.merge(all_epochs, window)

    all_epochs['Participant'] = 'participant012'

all_epochs = all_epochs.reset_index().iloc[:, 1:]
display(all_epochs)

all_epochs.to_csv("all_part_18_epochs/all_epochs_p012.csv") 




```

### Merge all the single participants into one big dataframe 

10 participants x 18 epochs = 180 datapoints with 230 features each 

```python

# get a list of all csv files for all participants 
participants = ['01', '02', '03', '04', '05', '07', '08', '10', '11', '12']
all_participants = [pd.read_csv(f'all_part_18_epochs/all_epochs_p0{p}.csv') for p in participants]

# iterate over the csv files and concatenate them into one big df with shape(180, 230) -- (180, 231) bc of participant info 
resting_state_complete = pd.DataFrame()
for part in all_participants: 
    resting_state_complete = pd.concat([resting_state_complete, part], axis=0)
    

resting_state_complete = resting_state_complete.reset_index().iloc[:, 2:]
print(resting_state_complete.shape)
display(resting_state_complete)

resting_state_complete.to_csv('resting_state_complete.csv') 



    
    
```
